### 7. Трехмерный график поверхности

Постройте 3D-график функции 
$$
z = \sin\left(\sqrt{x^2 + y^2}\right)
$$ 
для значений \( x, y \) в диапазоне \[-5, 5]\, используя библиотеку `mpl_toolkits.mplot3d`.

```python
x = np.linspace(-5, 5, 100)
y = x
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.set_title('Трехмерный график поверхности z')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')

ax.plot_surface(x,y,z, cmap='viridis')
plt.show()
```

### 5. Площадь под гиперболой

Найдите площадь под функцией:
$$ 
f(x) = \frac{1}{x}
$$
на интервале [1, 5].

```python
x = np.linspace(0.5,6,50)
y = 1/x

def f(x):
    return 1/x
# Используем метод трапеций для подсчета интеграла
def metodTrapez(f, a, b, n = 50):
    x = np.linspace(a, b, n + 1)
    y = f(x)
    dx = (b - a) / n
    integral = (dx / 2) * (y[0] + 2 * np.sum(y[1:n]) + y[n])
    return integral

plt.plot(x,y, c = 'r')
plt.fill_between(x,y, where=(x >= 1) & (x <= 5))

plt.title('Площадь под гиперболой в диапазоне [1,5]')
plt.xlabel('x')
plt.ylabel('y')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

integral = np.round(metodTrapez(f, 1, 5, 50), 4)
print(f'Интеграл в отрезке [1,5] равен: {integral}\n')

```

### 5. Коническая поверхность

Постройте конус:
$$
f(x,y) = \sqrt{x^2 + y^2}
$$
На интервале x,y [-5, 5]

```python
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
x, y = np.meshgrid(x, y)
z = np.sqrt(x**2 + y**2)

fig = plt.figure(figsize = (8,12))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap = 'viridis')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Коническая поверхность')
```
